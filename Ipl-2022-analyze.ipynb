DATA CLEANING 

```python
#data clean (remove null values) 
df = df.dropna() 
```

```python
#remove duplicates 
df = df.drop_duplicates()
```

## DATA ANALYZE  WITH VISUALIZATION ##

```python
#which team won the most match 
df['match_winner'].value_counts()
```

```python
plt.figure(figsize=(11,4))
sns.countplot(x=df["match_winner"], palette = ["#03045E", "#0077B6", "#00B4D8", "#90E0EF", "#CAF0F8"] )
plt.xticks(rotation=90)
plt.title("Total Wins by Team")
plt.savefig("images/MOST_MATCH_WINNER.png")
plt.show()
```

```python
#toss decision trend
df['toss_decision'].value_counts()
```

```python
sns.countplot(x="toss_decision", data=df ,  palette= ["#2ECC71" , "#D2B48C"])
plt.title("Toss Decisions")
plt.savefig("images/TOSS_DECISION.png")
plt.show()
```

```python
# toss winner and the match winner 
pd.crosstab(df['toss_winner'] ,  df['match_winner'])
```

```python
# top (player) performance  -  player of the match 
#head used to get top performers
df['player_of_the_match'].value_counts().head(10)
```

```python
top_player = df['player_of_the_match'].value_counts().head(10)
plt.figure(figsize=(14,5))
sns.barplot(y= top_player.values , x = top_player.index ,palette = ["#03045E", "#0077B6", "#00B4D8", "#90E0EF", "#CAF0F8"])
plt.title("Top Player of the Match Winners")
plt.savefig("images/PLAYER_OF_THE_MATCH.png")
plt.show()
```

```python
# which venue hosted most matches 
venue = df['venue'].value_counts()
```

```python
sns.barplot(x=venue.values , y = venue.index ,palette=sns.color_palette("Set2", 10))
plt.title("stadiums Hosting Most IPL Matches")
plt.savefig("images/VENUES.png")
plt.show()
```

```python
#correlation 
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.title("CORRELATION HEATMAP")
plt.savefig("images/CORRELATION.png")
plt.show()
```

```python
#how do teams win (runs vs wicket)
sns.countplot(x = df['won_by']  , palette = sns.color_palette("Spectral", 20))
plt.savefig("images/WON_BY.png")
plt.show()
```

```python
# 10 best bowling figures and bowlers
df['highest_wicket'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wicket'] = df['highest_wicket'].astype(int)
topBowlers = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending = False).head(10)
plt.figure(figsize=(10,5))
topBowlers.plot(kind = 'bar',color=sns.color_palette("Spectral"))
plt.savefig("images/BEST_BOWLERS.png")
plt.show()



```

```python
#who won the highest margin runs  

margin = df[df['won_by'] == 'Runs'].sort_values (by = 'margin' , ascending= False).head(10)[['match_winner' , 'margin']]
sns.barplot(x=margin['match_winner'] , y= margin['margin'] ,palette=sns.color_palette("husl", len(margin))) 
plt.xticks(rotation=90)
plt.title("WIN BY MARGINS")
plt.savefig("images/MARGIN_WIN.png")
plt.show()
```

```python
#which player has the highest  score  in the whole tournament 
highscorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending= False)
sns.barplot(y = highscorer.values , x = highscorer.index , palette = [
"#0B3C5D",  # very dark blue
"#0F4C81",
"#1F618D",
"#2E86C1",
"#3498DB",
"#5DADE2",
"#85C1E9",
   # very light blue
] ) 
plt.xticks(rotation = 90)
plt.savefig("images/TOP_SCORE_BATSMAN.png")
plt.show()
```

```python
#which player has the highest indivisual score  
df[df['highscore'] == df['highscore'].max()][['top_scorer' , 'highscore']  ]
```

```python
#which bowler had the best bowling figure 
bestBowling = df[df['highest_wicket'] == df['highest_wicket'].max()].sort_values(by='best_bowling_figure') [['best_bowling' , 'best_bowling_figure']]
bestBowling
```
